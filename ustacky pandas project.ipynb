# Dealing with DateTime Features

```python
# Convert 'Date' column datatype from str to datetime 
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#confirm change to datetime
print(df['Date'].dtype)
```

```python
#conversion of Time column to datetime
df['Time'] = pd.to_datetime(df['Time'])
```

```python
#confirm change to datetime
print(df['Time'].dtype)
```

```python
#extracting the days of the months of transcation
df['Day'] = df.Date.dt.day
df['Day']
```

```python
#extracting the months of transcation
df['Month'] = df.Date.dt.month
df['Month']
```

```python
#extracting the year of transaction
df['Year'] = df.Date.dt.year
df['Year']
```

```python
#extracting the hour of transaction
df['Hour'] = df.Time.dt.hour
df['Hour']
```

```python
#number of unique hours of transaction
df['Hour'].nunique()
```

```python
#array of unique hour sales
df['Hour'].unique()
```

# Aggregation with GroupBy

```python
#groupby city column on mean
df.groupby("City").mean()
```

```python
#groupby city column on mean
df.groupby("City").sum()
```

```python
#gross income of each city
city_gross=df.groupby("City")['gross income'].sum()
city_gross
```

```python
#The city with the maximum gross income
for i in city_gross:
    if i == city_gross.max():
        print('The city with the maximum gross income is:',city_gross.idxmax())
```
